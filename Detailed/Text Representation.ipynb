```python
def remove_html(text):
    
    return re.sub("<.*?>",'',text)
```

```python
df=df[:100]
```

```python
import string
exclude=string.punctuation
```

```python
# def remove_punc(text):
#     for char in exclude:
#         text=text.replace(char,'')
#     return text

def remove_punc(text):
    return text.translate(str.maketrans('','',exclude))
        
```

```python
from textblob import TextBlob

def correct_spelling(text):
    txtblb=TextBlob(text)
    return txtblb.correct().string
```

```python
# Remove stopwords
# import nltk
# from nltk.corpus import stopwords
lemmatizer=WordNetLemmatizer()
stemmer=PorterStemmer()
def remove_stopwords_stemming(text):
#     new_text=[lemmatizer.lemmatize(word) if word not in stopwords.words('english') else '' for word in nltk.word_tokenize(text)]
    new_text=[stemmer.stem(word) if word not in stopwords.words('english') else '' for word in nltk.word_tokenize(text)]
    
    return ' '.join(new_text)
        
        
```

```python
remove_stopwords("it is running")
```

```python
# Stemming
```

```python
def basic_text_preprocess(text):
    text=remove_html(text)
    text=remove_punc(text)
    text=correct_spelling(text)
    text=remove_stopwords_stemming(text)
    return text
```

```python
for i in range(len(df)):
    print(i)
    df.review[i]=basic_text_preprocess(df.review[i])
    print(df.review[i])
```

```python
import gensim
```

```python
from gensim.utils import simple_preprocess
```

```python
updated_review=[]
for ele in df.review:
    updated_review.append(simple_preprocess(ele))
```

```python
updated_review
```

```python
df.review[0]
```

```python
model=gensim.models.Word2Vec(window=10, min_count=2,vector_size=500)
```

```python
model.build_vocab(updated_review)
```

```python
model.epochs
```

```python
model.train(updated_review,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar("love")
```
